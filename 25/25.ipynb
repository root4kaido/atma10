```python
import pandas as pd
from pathlib import Path

import numpy as np
from tqdm import tqdm_notebook as tqdm
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
DATA_DIR = Path("/home/knikaido/work/atma10/data/")
```

```python
train = pd.read_csv(DATA_DIR / 'train.csv')
train['log_likes'] = np.log1p(train['likes'])
```

```python
oof_path = ['/home/knikaido/work/atma10/git/21/oof21.csv',
            '/home/knikaido/work/atma10/git/22/oof22.csv',
            '/home/knikaido/work/atma10/git/23/oof23.csv',
            '/home/knikaido/work/atma10/git/24/oof24.csv',
#             '/home/knikaido/work/atma10/git/26/oof26.csv',
#             '/home/knikaido/work/atma10/git/27/oof27.csv'
           ]
```

```python
oof_df = pd.DataFrame()
oof_df['log_likes'] = np.log1p(train['likes'])
for path in oof_path:
    oof_df[path.split('/')[-1][:-4]] = np.log1p(pd.read_csv(path))
oof_df
```

```python
y_pred = np.mean(oof_df.iloc[:, 1:], axis=1)
y_pred.shape
```

```python
np.sqrt(mean_squared_error(oof_df['log_likes'], y_pred))
```

```python
sub_path = ['/home/knikaido/work/atma10/git/21/sub21.csv',
            '/home/knikaido/work/atma10/git/22/sub22.csv',
            '/home/knikaido/work/atma10/git/23/sub23.csv',
            '/home/knikaido/work/atma10/git/24/sub24.csv',
            '/home/knikaido/work/atma10/git/26/sub26.csv',
#             '/home/knikaido/work/atma10/git/27/oof27.csv'
           ]
```

```python
sub_df = pd.DataFrame()
for path in sub_path:
    sub_df[path.split('/')[-1][:-4]] = np.log1p(pd.read_csv(path).values).reshape(-1)
sub_df
```

```python
y_pred = np.mean(sub_df.iloc[:, 1:], axis=1)
y_pred.shape
```

```python
sub = pd.read_csv(DATA_DIR / "atmacup10__sample_submission.csv")
sub["likes"] = np.expm1(y_pred)
sub
```

```python
sub.loc[sub.likes <= 0,"likes"] = 0
sub.to_csv('./sub_ansamble2.csv',index=False)
```
